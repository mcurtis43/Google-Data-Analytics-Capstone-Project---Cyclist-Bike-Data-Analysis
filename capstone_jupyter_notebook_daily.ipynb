```python
##This notebook creates a pivot table that links days of the week to the average rental lengh of a bike by either a member or casual renter
```

```python
import pandas as pd
import numpy as np
```

```python
dict_months = {

    0: 'January',
    1: 'February',
    2: 'March',
    3: 'April',
    4: 'May',
    5: 'June',
    6: 'July',
    7: 'August',
    8: 'September',
    9: 'October',
    10: 'November',
    11: 'December'
    
}

## creates a dictionary that will be referenced later to pull in month names for data frames and monthly data analysis
```

```python
df_jan = pd.read_csv('jan_2025_raw_trip_data.csv')
df_feb = pd.read_csv('feb_2025_raw_trip_data.csv')
df_mar = pd.read_csv('mar_2025_raw_trip_data.csv')
df_apr = pd.read_csv('apr_2025_raw_trip_data.csv')
df_may = pd.read_csv('may_2025_raw_trip_data.csv')
df_jun = pd.read_csv('jun_2025_raw_trip_data.csv')
df_jul = pd.read_csv('jul_2025_raw_trip_data.csv')
df_aug = pd.read_csv('aug_2025_raw_trip_data.csv')
df_sep = pd.read_csv('sep_2025_raw_trip_data.csv')
df_oct = pd.read_csv('oct_2025_raw_trip_data.csv')
df_nov = pd.read_csv('nov_2025_raw_trip_data.csv')
df_dec = pd.read_csv('dec_2025_raw_trip_data.csv')

## import all csv files as dataframes
```

```python
df_list = [df_jan, df_feb, df_mar, df_apr, df_may, df_jun, df_jul, df_aug, df_sep, df_oct, df_nov, df_dec]

##creates a list of data frames to pefrom opperations on 
```

```python
list_started_at = []
list_ended_at = []

for i in range(len(df_list)):
    columns_to_check = ['started_at', 'ended_at']
    list_started_at.append(str(df_list[i][columns_to_check[0]].isna().sum()))
    list_ended_at.append(str(df_list[i][columns_to_check[1]].isna().sum()))
print(list_started_at)
print(list_ended_at)

##Checks to make sure there are no na values in the started at or ended at columns of the df
```

```python
list_classic_electric = []

for i in range(len(df_list)):
    classic_rideable_values = (df_list[i]['rideable_type'] == 'classic_bike' ).sum()
    electric_rideable_values = (df_list[i]['rideable_type'] == 'electric_bike' ).sum()
    bad_rideable_values = (len(df_list[i]) - classic_rideable_values - electric_rideable_values)
    list_classic_electric.append(str(bad_rideable_values))
print(list_classic_electric)

##Makes sure that the total number of rentals is equal to the mechanical rentals plus the electrical rentals
```

```python
list_member_casual = []

for i in range(len(df_list)):
    member_values = (df_list[i]['member_casual'] == 'member' ).sum()
    casual_values = (df_list[i]['member_casual'] == 'casual' ).sum()
    bad_member_values = (len(df_list[i]) - member_values - casual_values)
    list_member_casual.append(str(bad_member_values))
print(list_member_casual)

##makes sure that the total users is equal to the number of casual members plus the number of members
```

```python
for i in range(len(df_list)):
    df_list[i]['started_at'] = pd.to_datetime(df_list[i]['started_at'])
    df_list[i]['ended_at'] = pd.to_datetime(df_list[i]['ended_at'])

    ##converts the started at and ended at columns from a string value to a date time value
```

```python
for i in range(len(df_list)):
    df_list[i]['duration'] = df_list[i]['ended_at'] - df_list[i]['started_at']
    
    ##adds a column to the df, duration, which measures how long a bike was rented for
```

```python
for i in range(len(df_list)):
    df_list[i]['duration_mins'] = (df_list[i]['duration'].dt.total_seconds()/60).round(2)

    ##converts the date time format into a total minutes float value
```

```python
list_illogical_duration = []

for i in range(len(df_list)):
    illoical_duration = (df_list[i]['duration'] < pd.Timedelta(0)).sum()
    list_illogical_duration.append(str(illoical_duration))

print(list_illogical_duration)

## checks to see if any ride durations are illogical, e.g. returned before the ride even started. December has 29 of these values.
```

```python
df_dec = df_dec[df_dec['duration'] >= pd.Timedelta(0)]

##reduces december to only have rows with a duration > 0
```

```python
df_list = [df_jan, df_feb, df_mar, df_apr, df_may, df_jun, df_jul, df_aug, df_sep, df_oct, df_nov, df_dec]
list_illogical_duration_01 = []

for i in range(len(df_list)):
    illoical_duration_01 = (df_list[i]['duration'] < pd.Timedelta(0)).sum()
    list_illogical_duration_01.append(str(illoical_duration_01))

print(list_illogical_duration_01)

## rechecks the above conditions to make sure there are no rows with duration < 0. This is now fixed
```

```python
for i in range(len(df_list)):
    df_list[i]['week_day'] = df_list[i]['started_at'].dt.day_name()

    ##add column with day of week based on column start date
```

```python
output_list_january_mem = []
output_list_january_cas = []
output_list_february_mem = []
output_list_february_cas = []
output_list_march_mem = []
output_list_march_cas = []
output_list_april_mem = []
output_list_april_cas = []
output_list_may_mem = []
output_list_may_cas = []
output_list_june_mem = []
output_list_june_cas = []
output_list_july_mem = []
output_list_july_cas = []
output_list_august_mem = []
output_list_august_cas = []
output_list_september_mem = []
output_list_september_cas = []
output_list_october_mem = []
output_list_october_cas = []
output_list_november_mem = []
output_list_november_cas = []
output_list_december_mem = []
output_list_december_cas = []

list_output_cas = [
output_list_january_cas,
output_list_february_cas,
output_list_march_cas,
output_list_april_cas,
output_list_may_cas,
output_list_june_cas,
output_list_july_cas,
output_list_august_cas,
output_list_september_cas,
output_list_october_cas,
output_list_november_cas,
output_list_december_cas,

]

list_output_mem = [

output_list_january_mem,
output_list_february_mem,
output_list_march_mem,
output_list_april_mem,
output_list_may_mem,
output_list_june_mem,
output_list_july_mem,
output_list_august_mem,
output_list_september_mem,
output_list_october_mem,
output_list_november_mem,
output_list_december_mem,

]

day_of_week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

for i in range(7):
    list_output_mem[i].append(day_of_week[i])
    list_output_cas[i].append(day_of_week[i])
    list_output_mem[i].append('member')
    list_output_cas[i].append('casual')
    print(list_output_mem[i])
    print(list_output_cas[i])

    ##create output list used to start building pivot table manually
```

```python
df_list = [df_jan, df_feb, df_mar, df_apr, df_may, df_jun, df_jul, df_aug, df_sep, df_oct, df_nov, df_dec]
day_of_week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

for i in range(len(day_of_week)):
    for j in range(len(df_list)):
        df_mem = df_list[j][df_list[j]['member_casual'] == 'member']
        df_mem_wk = df_mem[df_mem['week_day'] == day_of_week[i]]
        list_output_mem[i].append(str(df_mem_wk['duration_mins'].mean().round(2)))

for i in range(len(day_of_week)):
    for j in range(len(df_list)):
        df_cas = df_list[j][df_list[j]['member_casual'] == 'casual']
        df_cas_wk = df_cas[df_cas['week_day'] == day_of_week[i]]
        list_output_cas[i].append(str(df_cas_wk['duration_mins'].mean().round(2)))

## uses two double for loops to disect the dataframes and return a count value. This is performed once to disect by day of the week and once to disect by member/casual markers.
```

```python
for k in range(7):
    print(list_output_mem[k])
    print(list_output_cas[k])
```

```python
output_columns = ['day', 'member_casual', 'jan_avg', 'feb_avg', 'mar_avg', 'apr_avg', 'may_avg', 'jun_avg', 'july_avg', 'aug_avg', 'sep_avg', 'oct_avg',
                  'nov_avg', 'dec_avg']
df_output_weekly = pd.DataFrame(columns=output_columns)

for i in range(7):
    df_output_weekly.loc[len(df_output_weekly)] = list_output_mem[i]
    df_output_weekly.loc[len(df_output_weekly)] = list_output_cas[i]
df_output_weekly.head(24)

## creates the desired pivot table for analysis.
```

```python
df_output_weekly.to_csv('cleaned_weekly_data.csv', index = 'False')
```
